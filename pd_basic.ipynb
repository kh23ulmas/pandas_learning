```python
import pandas as pd
df_students = pd.read_csv('data/students.csv', sep=';')
ans = df_students[' Subject'].unique()
print(len(ans))

print(df_students[' grade'].mean())
```

```python
df_students['EurGrad'] = pd.Series(map(\
    lambda x: \
        'A' if x >= 90 and x <= 100 else \
        'B' if x >= 85 and x <= 89 else \
        'C' if x >= 75 and x <= 84 else \
        'D' if x >= 65 and x <= 74 else \
        'E' if x >= 60 and x <= 64 else \
        'F', df_students[' grade']))
print(*df_students['EurGrad'])
```
