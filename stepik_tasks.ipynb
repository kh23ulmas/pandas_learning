```python
taxi.rename(columns={'pcp 01':'pcp_01', 'pcp 06':'pcp_06', 'pcp 24':'pcp_24'}, inplace=True)
taxi
```

```python
taxi['pickups'].sum()
```

```python
taxi.groupby(['borough']).agg({'pickups':'sum'}).sort_values('pickups', ascending=False)
min_pickups = taxi.groupby(['borough']).agg({'pickups':'sum'}).idxmin()
min_pickups
```

```python
# Для каждого района посчитайте число поездок по месяцам. Отсортируйте полученные значения по убыванию и сохраните результирующий датафрейм в pickups_by_mon_bor.

# Обратите внимание, что итоговый датасет должен состоять из 3-х колонок: pickup_month, borough, pickups.
def make_zero(ds):
    print(ds[0])
    for i in range(len(ds)):
        ds[i] = 0
    return ds
pickups_by_mon_bor = (taxi
                        .groupby(['pickup_month', 'borough'], as_index=False)
                        .agg({'pickups':'sum'})
                        .sort_values('pickups', ascending=False))
# type(pickups_by_mon_bor['pickups'])

print(make_zero(pickups_by_mon_bor['pickups']))
```
