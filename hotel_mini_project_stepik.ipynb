2. Пользователи из каких стран совершили наибольшее число успешных бронирований? 
Бронирование считается успешным, если в дальнейшем оно не было отменено (переменная is_canceled). В качестве ответа выберите страны, входящие в топ-5.

```python
ans = (bookings
       .loc[bookings['is_canceled']==0]
       .groupby('country')
       .count()
       .sort_values('hotel', ascending=False))
ans.head()
# version 2
ans_2 = (bookings
         .loc[bookings['is_canceled']==0]
         .country
         .value_counts())
ans_2.head()
```

3. На сколько ночей (stays_total_nights)  в среднем бронируют отели типа City Hotel? Resort Hotel? Запишите полученные значения в пропуски с точностью до 2 знаков после точки.

```python
bookings.head()
# stays_total_nights
ans = (bookings
        .loc[(bookings.hotel == 'City Hotel') | (bookings.hotel =='Resort Hotel')]
        .groupby('hotel')
        .mean('stays_total_nights'))
ans['stays_total_nights'] = round(ans['stays_total_nights'], 2)
ans['stays_total_nights']

#version 2 karpov

ans_2 = bookings.groupby('hotel').stays_total_nights.mean().round(2)
ans_2
```
