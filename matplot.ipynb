```python
class MyNum:
    def __init__(self, num):
        self.num = num
    def __add__(self, n):
        return MyNum(self.num + n)
    
    def __str__(self) -> str:
        return f"number {self.num}"
    def get_val(self):
        return self.num
```

```python
n = MyNum(6)
n.get_val()
print(type(n))
n += 5
print(type(n))
```
